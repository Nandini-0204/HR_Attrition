# Visualiztion of Employee Distribution

```python
dg=df.groupby('Department')['JobRole'].count()
```

```python
dg
```

```python
dg=dg.reset_index()
sb.barplot(x=dg.Department,y=dg.JobRole , palette='deep' )
plt.xticks(rotation=45)
plt.xlabel("Department")       
plt.ylabel("JobRole")        
plt.title("Job Roles by Department")
plt.show()
```

```python
sb.histplot(df['Age'],bins=[0,10,20,30,40,50,60,70] ,color='darkgreen')
plt.xticks(rotation=45)
plt.xlabel("Age Groups")       
plt.ylabel("Count")        
plt.title("Age Distribution")  
plt.show()
```

```python
sb.histplot(df['MonthlyIncome'],bins=[0,2500,5000,7500,10000,12500,15000,17500,20000] , color='DarkGreen')
plt.xlabel("Monthly Income")
plt.ylabel("Count")
plt.title("Distribution by Monthly Income")
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
sb.boxplot(x='Attrition',y='JobSatisfaction',data=df , palette='cool')
plt.xticks(rotation=45)
plt.xlabel("Attrition")
plt.ylabel("Job Satisfaction")
plt.title("Job Satisfaction by Attrition")

plt.subplot(1,3,2)
sb.boxplot(x='Attrition',y='MonthlyIncome',data=df , palette='cool')
plt.xticks(rotation=45)
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.title("Monthly Income by Attrition")



plt.subplot(1,3,3)
sb.boxplot(x='Attrition',y='PerformanceRating',data=df , palette='cool')
plt.xticks(rotation=45)
plt.xlabel("Attrition")
plt.ylabel("Performance Rating")
plt.title("Performance Rating by Attrition")
plt.show()
```

```python
plt.figure(figsize=(18,20))
plt.subplot(2,2,1)
sb.countplot(x=df.Gender,hue=df.Attrition , palette='Greens')
plt.xticks(rotation=45)
plt.title("Gender Distribution by Attrition")

plt.subplot(2,2,2)
sb.countplot(x=df.Age , hue=df.Attrition , palette='Greens')
plt.xticks(rotation=90)
plt.title("Age Distribution by Attrition")

plt.subplot(2,2,3)
sb.countplot(x=df.Department , hue=df.Attrition , palette='Greens')
plt.xticks(rotation=45)
plt.title("Departmnet Distributiob by Attrition")

plt.subplot(2,2,4)
sb.countplot(x=df.EducationField , hue=df.Attrition , palette='Greens')
plt.xticks(rotation=45)
plt.title("Educational Field Distribution by Attrition")
plt.show()
```
